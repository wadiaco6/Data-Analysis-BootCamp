# Data Visualization

Create a scattered plot for release year and title, and provide some observation in your own words

```python
# Importing the required liberary
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creating a Scattered plot
title = ['Dick Johnson Is Dead', 'Blood & Water', 'Ganglands', ' Jailbirds New Orleans', 'Kota Factory', 'Zodiac', 'Zombie Dumb', 'Zombieland', 'Zoom', 'zubaan']
release_year = ['2020', '2021', '2021', '2021', '2021','2007', '2018', '2006', '2009', '2015']

         
# Draw the scatter plot
plt.scatter(release_year,title,c='red', marker='*')

# adding x label
plt.xlabel('This is the x-axis')

# adding x label
plt.ylabel('This is the y-axis')

# Adding the title
plt.title('This is the title of the scatter plot')


# show the graph
plt.show()
```
